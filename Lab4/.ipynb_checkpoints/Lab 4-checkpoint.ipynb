# Economic Success of Various Astrological Signs

The goal of this research is to determine if Astrological Signs play any role in the economical success of a person.  In the case of this study, we will assume that economic success is defined as some combination of income, hours worked per week, and a ratio of hours worked to income (checking a persons hourly wage).  To find this information, the mean, median, and histogram distributions will be created for each of these statistics.


All data used in this lab comes from GSS DataExplorer, a part of the National Opinon Research Center, hosted by the University of Chicago.  GSS can be found at https://gssdataexplorer.norc.org

## Question
Do astrological signs affect how economically successful somebody is in the long run?

## Data Collection

```python
import csv
import statistics
import matplotlib.pyplot as plt
import scipy.stats as st
import random as rand
%matplotlib inline
    
f = open('astro2.csv', 'r')
csv_f = csv.reader(f)

astros = {}

#first collect data
for row in csv_f :
    #create the dictionary to store data
    if row[8] != "Respondents astrological sign" and row[8] != "":
        if row[8] not in astros:
            astros[row[8]] = {"hours": [], "income": []}  
        
        #how much the respondant works
        astros[row[8]]["hours"].append(row[0])
        
        #income of the respondant
        astros[row[8]]["income"].append(row[3])                
```

## Income

```python
#lets get some income stats
print ("average income for each astrological signs")
for key in astros:
    moneycount = 0
    for i in astros[key]["income"]:
        moneycount += int(i)
    print(key, ": $", round(moneycount/len(astros[key]["income"])))
```

There is an interesting difference in average income between a couple astrological signs.  Most notably, there is almost a 15000 dollar difference between Taurus and Capricorns. Another high income sign appears to be Gemini, with a mean of $29339.  Histograms will be produced for the highest and lowest astrological signs in order to investigate this further.

```python
#Taurus
plt.xlabel("Income")
plt.ylabel("Number of Respondants")
f = plt.figure(1)
int_list = []
plt.title("Taurus")
for i in astros["Taurus"]["income"]:
    int_list.append(int(i))
plt.hist(int_list)

plt.xlabel("Income")
plt.ylabel("Number of Respondants")
#Capricorn
f = plt.figure(2)
int_list = []
plt.title("Capricorn")
for i in astros["Capricorn"]["income"]:
    int_list.append(int(i))
plt.hist(int_list)

plt.xlabel("Income")
plt.ylabel("Number of Respondants")
#Gemini
f = plt.figure(3)
int_list = []
plt.title("Gemini")
for i in astros["Gemini"]["income"]:
    int_list.append(int(i))
plt.hist(int_list)

plt.xlabel("Income")
plt.ylabel("Number of Respondants")
#Libra
f = plt.figure(4)
int_list = []
plt.title("Libra")
for i in astros["Libra"]["income"]:
    int_list.append(int(i))
plt.hist(int_list)

plt.xlabel("Income")
plt.ylabel("Number of Respondants")
#Leo
f = plt.figure(5)
int_list = []
plt.title("Leo")
for i in astros["Leo"]["income"]:
    int_list.append(int(i))
plt.hist(int_list)
```

It appears that outliers are having a large impact on the calculated averages for each astrological signs. In order to confirm any findings, investigating medians could provide insight see if they is any difference in income between astrological signs.

```python
#calculate median
#lets get some income stats
print ("Median income for each astrological sign")
for key in astros:
    int_list = []
    for i in astros[key]["income"]:
        int_list.append(int(i))
    print(key + " Median: ", statistics.median(int_list))
```

Calculating medians, it appears that yes, there is some difference in income of the different astrological signs.  From this data it appears that Gemini, one of the originally high mean signs, and Sagittarius, a slightly above average sign have the highest medians.  This descrepincy within data leads me to beileve that there are outside factors, our simply outliers, which are affecting these calculations.  While it does appear that, at least from an income standpoint, different astrological signs do appear to be more successful than other, the validity of this conclusion can not be fully determined at this time.  In order to continue our experimentation, other factors will be observed.

Next hours worked per week will be examined for each sign

## Hours Worked per Week

```python
#lets get some work stats
print ("Average hours worked per week by each Astrological Sign")
for key in astros:
    workstuff = 0
    for i in astros[key]["hours"]:
            workstuff += int(i)  
    print(key, ":", round(workstuff/len(astros[key]["hours"]), 2))
```

There appears to be a slight difference in hours worked per week between various astrological signs. Histograms will be used to examine the highest and lowest signs.

```python
counter = 1

for key in astros:
    f = plt.figure(counter)
    int_list = []
    plt.title(key)
    for i in astros[key]["hours"]:
        int_list.append(int(i))
    plt.xlabel("Hours worked")
    plt.ylabel("Number of respondants")
    plt.hist(int_list)
    counter += 1
```

From these histograms, we can determine a couple of pieces of information.  As stated by the calculated means, it is confirmed that more Virgos work more hours than other astrological signs.  It can also be seen that while it doesn't have the lowest overall mean hours, the Tarus histogram seems to show that more Tarus' work less than other signs, however increased working in the higher hours compared to other signs helps it's mean even out.  Libras, tied for last with Pieces, surpusringly show a very samll amount of people working low hours compared to other signs, however their lack of laborers in the "20-30" hours region as well as higher hours lead to a low mean.

To confirm that there are indeed differences in hours worked for various astrological signs,  the median hours worked for each sign will be calcuated

```python
for key in astros:
    int_list = []
    for i in astros[key]["hours"]:
        int_list.append(int(i))
    print(key + " Median: ", statistics.median(int_list))
```

These new median calculations appear to contradict our findings.  Despite differnet mean hours worked and differnt histograms for each sign, the median values are all the same.  This could conclude that outliers within our dataset are skewing the caculated means, and that in reality, astrological signs have little to nothing to do with total hours worked.  In each case, the majority of workers worked an average work week of 40 hours.  While their may be slight differences in how the workforce of an astrological sign is distributed throughout the workforce in terms of hours, there is no conclusive evidence to state that people of certain astrological signs work signifagently more.

However, a topic that could be examined to check for economic success is how much each sign works for their money.  In order to do this, we will plot hours worked against respondant income for each astrological sign.  A steeper slope would indicate getting paid more for each hour, while a lower slope would indicate getting paid less.

## Income compared to hours worked

```python
counter = 1
for key in astros:
    hours = []
    money = []
    f = plt.figure(counter)
    plt.title(key)
    plt.ylabel("Income in Constant Dollars")
    plt.xlabel("Hours Worked Last Week")
    for i in astros[key]["income"]:
            money.append(int(i))  
    for i in astros[key]["hours"]:
            hours.append(int(i))  
    plt.scatter(hours, money)
    counter += 1
```

While these scatterplots definitley show different distributions for each astrological sign, looking at the slopes as raw values from linear regression would be more conclusive.

```python
for key in astros:
    hours = []
    money = []
    for i in astros[key]["income"]:
            money.append(int(i))  
    for i in astros[key]["hours"]:
            hours.append(int(i))  
    slope, intercept, r_value, p_value, std_err = st.linregress(hours,money)
    print(key + " slope", slope)
```

In this analysis, Leo and Taurus, two of the highest calculated mean signs, show the highest slope, indicating getting paid the most per hour.  This leads to the conclusion that these two signs may in fact be more successful than others. 

From these calculations, the most efficent money makers (according to the calculated slope) are in order, Leo, Taurus, Aquarius, Gemini, Sagittarius, Scorpio, Cancer, Aries, Libra, Capricorn, Pisces, Virgo.

To confirm this once more, an average wage per week (estimated) will be calcuated from each respondant.  

```python
for key in astros:
    hours = []
    money = []
    avg = []
    for i in astros[key]["income"]:
            money.append(int(i))  
    for i in astros[key]["hours"]:
            hours.append(int(i))  
    for i in range(len(hours)):
        avg.append(money[i]/hours[i])
    print(key + " average wage/week:", statistics.mean(avg))
```

Calculating the average wage per week for each sign has returned confusing data yet interp.  In the original scatterplots that Leo and Taurus were the most "efficent" at generating income, however in these individual calcuations, they appear to be close to average.  Further investigating shows that most signs cluster from around $500 - $700 dollars per week. Piseces was one of the only substaintially higher mean wage per week, and looking at the scatterplot, the substaintial portions of outliers seem to be skewing the results.  Due to the odd distribution of Piseces, this data is disregarded and the original slope calcuations will be primarily used.

Sagittarius and , however, also preformed well in both the calculated slope and the calcuated mean.  This leads me to believe that Sagitarius may be exceptionally efficent workers compared to the rest of the set.

## Results

The investigation of the question, "Does Astrological sign play any role in economic success?", provided very interesting data.  Overall this question was answered with surprising amounts of success, as explained by the summary below.

Investigating mean income for each sign, the most successful signs appear to be Taurus, Gemini, and Leo.  These signs had significantly higher mean incomes compared to the other signs, and show a possible correlation between astrological sign and income.

These original mean income calculations were found to be skewed by substantial outliers, lowering their accuracy.  However, this data proved that there was some difference in income between astrological signs and allowed us to propose new investigative techniques such as median and histograms.

Examining the histograms of the most dramatic income discrepancies, it was clear that there were large differences in the distribution of income from various signs.  Large outliers that had effects on mean calculations could be seen as well.  In order to better represent this data, medians were calculated.

Median calculations provided different results than the mean calculations.  In terms of median, of the original successful mean income signs, only Gemini remained substantially more successful.  Sagittarius was also a well performing sign, however, Sagittarius performed poorly in the original mean.

From these calculations it was determined that while minimal, there do appear to be slight changes in income over various signs.  It was difficult to determine a "best" sign, but astrological sign does seem to play a role in overall income.

Next hours worked per week were compared.  While mean appeared to show differences originally, and histograms show varied distributions throughout each sign, the median proved that the different means were most likely due to outliers, since all signs had a median value of 40 hours per week.  While this did show that there was a change in hours worked per week by each sign, it was not substantial enough to work into the final decision.

After investigating both hours worked and overall income, the combination of these two variables, wage per week was investigated as well.  First, the two variables were plotted in scatter plots, showing the dataset for each sign as a whole.  These scatterplots again showed different distributions for each sign, and showed different apparent slopes for each variable.  Slopes based on linear regression were then calculated to determine quantitatively how each sign preformed in this category.  In these calculations, a higher slope would constitute a higher wage per hour worked, and a lower slope would mean lower pay.  From these results it was found that Leo, Taurus, Aquarius, Gemini, and Sagittarius had the highest overall slope.

To confirm these findings, wage per week were calculated for each respondent and averaged for each sign.  From this data, the only overlap that was found with the slope data was Sagittarius.  This lead to the conclusion that based on performing well in both tests, Sagittarius was one of the most efficient income earners, with both a high slope and high mean wage/week.  Other signs were found as well but were eliminated due to high outliers within the dataset.
